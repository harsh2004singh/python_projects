```python
from PIL import Image
```

```python
js=qr.QRCode(version=1,error_correction=qr.constants.ERROR_CORRECT_H,box_size=10,border=4,)
```

```python
js.add_data("linkedin.com/in/harsh-singh-337697232")
```

```python
js.make(fit=True)
```

```python
img=js.make_image(fill_color="red",back_color="blue")
```

```python
img.save("linkedin profile.png")
```
