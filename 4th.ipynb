```python
import speech_recognition as sr
import pyttsx3
import datetime
import webbrowser
import pyjokes
import time


def sptext():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)

        try:
            print("Recognizing...")
            data = recognizer.recognize_google(audio)
            print("You said:", data)
            return data
        except sr.UnknownValueError:
            print("Not understood")
            return ""


def speechtx(X):
    engine = pyttsx3.init()
    voices = engine.getProperty("voices")
    engine.setProperty("voice", voices[0].id)
    engine.setProperty("rate", 150)
    engine.say(X)
    engine.runAndWait()


if __name__ == "__main__":

    print("Say 'hey jarvis' to start")

    # üîÅ wake word loop
    while True:
        wake = sptext().lower()
        if "hey jarvis" in wake:
            speechtx("Hello, I am Jarvis")
            break

    # üîÅ command loop
    while True:
        dt1 = sptext().lower()

        if not dt1:
            continue

        if "your name" in dt1:
            speechtx("My name is Jarvis")

        elif "old are you" in dt1:
            speechtx("I am 21 years old")

        elif "developer" in dt1:
            speechtx("My developer is Harsh Singh")

        elif "how are you" in dt1:
            speechtx("I am fine, what about you")

        elif "now time" in dt1:
            speechtx(datetime.datetime.now().strftime("%I:%M %p"))

        elif "today date" in dt1:
            speechtx(str(datetime.date.today()))

        elif "what day" in dt1:
            speechtx(datetime.date.today().strftime("%A"))

        elif "youtube" in dt1:
            webbrowser.open("https://www.youtube.com/")

        elif "chat gpt" in dt1:
            webbrowser.open("https://chatgpt.com/")

        elif "instagram" in dt1:
            webbrowser.open("https://www.instagram.com/")

        elif "jokes" in dt1:
            joke = pyjokes.get_joke(language="en", category="neutral")
            print(joke)
            speechtx(joke)

        elif "exit" in dt1 or "bye" in dt1:
            speechtx("Thank you, goodbye")
            break

        time.sleep(1)
         
    else:
        print("sorry i can't hear you")
```
